```python
import pandas as pd
import numpy as np
```

```python
p = pd.read_csv("./phishtank-data.csv")
o = pd.read_csv("./openphish-data.csv")
b = pd.read_csv("./benign-data.csv")
k = pd.read_csv("./kernun-data.csv")
```

```python
ok = b.append(k, sort=True)
bad = p.append(o, sort=True)
```

### INPUT_TAG

```python
input_tag = pd.concat([ok.input_tag_value.value_counts(), bad.input_tag_value.value_counts()], axis=1)
input_tag.columns = ['ok', 'bad']
input_tag.fillna(0, inplace=True)
input_tag['ok_ratio'] = input_tag.ok / sum(input_tag.ok)
input_tag['bad_ratio'] = input_tag.bad / sum(input_tag.bad)
input_tag
```
